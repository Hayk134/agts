# 2. Решение задач робота и кибериммунности
В данном модуле вам предстоит решить задачу робота (5) и задачу кибериммунности CybZ_4 сначала в ЦД, а затем и на физической версии робота.

```python
# ЭТОТ БЛОК КОДА МОЖНО И НУЖНО МОДИФИЦИРОВАТЬ
# Как обычно, подсказки - в импортах

from src.libs.AGTSWrapTwin.Modules.Handler import MissionHandler, TrustedHandler
import time
import crc8
import datetime
import threading


class UserMissionHandler(MissionHandler):
    @staticmethod
    def config_cyber_obstacles():
        # Сервисная функция, в которой вы ОБЯЗАНЫ передать конфигурацию киберпрепятствий.
        # Изменять можно только сами значения флагов - если функция вернёт что-то кроме данного объекта, возникнет ошибка.

        return {
            "CybP_01": True,
            "CybP_02": True,
            "CybP_03": True,
            "CybP_04": True,
            "CybZ_01": True,
            "CybZ_02": True,
            "CybZ_03": True,
            "CybZ_04": True,
        }

    def mission_code(self):
        self.lg.log("Тестовое сообщение блока кода для выполнения заезда")
        self.ut_counter = 0
        self.res = [0, 0, 0]
        self.cell_utechka = 0

            
        self.colors = [1,2,3,4]
        self.colors_temptures = {
            9000 : [1,3],
            2000 : [2,4]
        }
        self.count_shkaf = 3
        self.cur_count_shkaf = 0
        

        self.errors = {
            (1,3,2) :"ВЗЛОМ ШЛЮЗОВ БЕЗОПАСНОСТI",
            (1,3,4) :"САБОТАЖ СИСТЕМЫ ОХЛАЖДЕНИЯ",
            (1,4,2) :"ЭНЕРГЕТИЧЕСКИЙ ШАНТАЖ",
            (3,1,2) :"ЛОВУШКА НА ПУТИ",
            (3,1,4) :"ОТРАВЛЕННЫЕ ДАННЫЕ",
            (3,2,4) :"ОБМАНЧИВОЕ ЗАТИШЬЕ",
            (3,4,2) :"НЕОЖИДАННАЯ АТАКА",
            (2,1,3) :"КОД ЗАМЕДЛЕННОГО ДЕЙСТВИЯ",
            (2,3,1) :"ПИРРОВА ПОБЕДА",
            (4,1,3) :"БУНТ МАШИН",    
        }
        
        cord_list = [

            (3.0, 0.6),
            (3.0, 0.7),
            (3.0, 2.6),
            (2.6, 2.6),
            (2.6, 3.3),
            (2.6, 2.6),
            (1.4, 2.6),
            (1.4, 0.6),
            (0.6, 0.6),
            (1.4, 0.6),
            (1.4, 2.2),

            (0.6, 2.2),
            (1.0, 2.2),
            (0.6, 1.8),
            (1.0, 1.8),

            (1.4, 3.0),
            (0.6, 3.0),
            (1.0, 3.0),
            (1.4, 3.0),
            (1.4, 1.0),
            (2.2, 1.0),
            (2.2, 1.8),
            (3.0, 2.6),
            (3.2, 2.2),
            (3.4, 2.2)

        ]

        vzyat1 = (2.6, 3.3)
        brosit1 = (0.6, 0.6)
        ochistka1 = (2.2, 1.8)
        shkaf2 =  (1.4, 3.0)
        shkaf1 = (1.0, 3.0)
        shkaf0 =  (0.6, 3.0)
        utechka1 =   (0.6, 2.2)
        utechka2 = (1.0, 2.2)
        utechka3 = (0.6, 1.8)
        utechka4 = (1.0, 1.8)

        self.idx_to_cord = {
        shkaf0 : 0,
        shkaf1 : 1,
        shkaf2 : 2
            }

        action = {vzyat1: self.vzyat,
                  brosit1: self.brosit,
                  ochistka1: self.ochistka,
                  utechka1: self.utechka,
                  utechka2: self.utechka,
                  utechka3: self.utechka,
                  utechka4: self.utechka,
                  shkaf2: self.shkaf,
                  shkaf1: self.shkaf,
                  shkaf0: self.shkaf,
                  }

        for cord in cord_list:
            x, y = cord

            if cord in action:
                do = action[cord]
            else:
                do = self.move

            do(x, y)

    def move(self, x, y):
        self.ap_hook.do_rotate({"x": x, "y": y})
        self.ap_hook.do_move({"x": x, "y": y}, autorotate=False)

    def vzyat(self, x, y):

        self.ap_hook.do_rotate({"x": x, "y": y})
        self.ap_hook.do_move({"x": x, "y": y}, autorotate=False)
        self.ap_hook.do_grip(True)
        self.lg.log("Робот соверщил захват")

    def brosit(self, x, y):

        self.ap_hook.do_rotate({"x": x, "y": y})
        self.ap_hook.do_move({"x": x, "y": y}, autorotate=False)
        self.ap_hook.do_grip(False)
        self.lg.log("Робот соверщил сбросс")

    def ochistka(self, x, y):

        self.ap_hook.do_rotate({"x": x, "y": y})
        self.ap_hook.do_move({"x": x, "y": y}, autorotate=False)
        self.set_begin_cleaning()
        self.do_wait(strategy="time", duration=10)
        self.lg.log("Очистка произошла успешно")


    def shkaf(self,x,y):
            self.cur_count_shkaf += 1
    
            self.ap_hook.do_move({"x": x, "y": y}, autorotate=False)
    
            colid = None
            idx = self.idx_to_cord.get((x,y))
            data = self.get_current_control_status()
            colid = data['control_color']
            self.res[idx] = colid
            self.lg.log(f'{self.res[idx]}')
            if self.cur_count_shkaf == self.count_shkaf:
                res = tuple(self.res)
                oshibka = self.errors.get(res, "Не получилось")
                if oshibka == "Не получилось":
                    res = self.get_res_with_reserve(list(self.res))
                    oshibka = self.errors.get(res)
                self.lg.log(f'{oshibka}')

    def get_res_with_reserve(self, colors : list) -> tuple:
        for idx,color in enumerate(colors):
            if color not in self.colors:
                reserve_color_temputure = self.get_message_from_trusted_module()[0]
                possible_colors = self.colors_temptures.get(reserve_color_temputure,[])
                for possible_color in possible_colors:
                    if possible_color not in colors:
                        colors[idx] = possible_color
                        break
        return tuple(colors)

    def utechka(self,x,y):
        self.cell_utechka += 1
        self.ap_hook.do_rotate({"x": x, "y": y})
        self.ap_hook.do_move({"x": x, "y": y}, autorotate=False)
        ut = self.get_current_cell_pollution_status()
        self.ut_counter = (self.ut_counter + 1) if ut else self.ut_counter
        if ut:
            self.lg.log(f'зона {self.ut_counter}.{ut.get('pollution_status')}')
        if self.ut_counter == 4:
            self.lg.log('утечка')
            time.sleep(1)
            b = self.get_message_from_trusted_module() 
            self.lg.log(f'утечка в месте{b}')



    # Вы можете добавить ваши собственные функции, которые будут вызываться mission_code (подсказка - если вы вынесете функционал в отдельную функцию, его проще будет оценить в случае разногласий).
    # Автоматически вызывается только функция mission_code (один раз при вызове start_module())


class UserTrustedHandler(TrustedHandler):
    def __init__(self, context):
        super().__init__(context)
        self.CybP_02_status = 0
        self.short_massage = None
        self.ap_hash: dict
        self.CybP_01_act = False

        self.cell_raduis = 0.2
        self.cell_centre_ration = 0.25
        self.cell_centre_raduis = self.cell_raduis * self.cell_centre_ration

        self.pos_x = 0
        self.pos_y = 0
        self.evacuation = False
        self.CybZ_02 = False

        self.next_t = time.time()
        self.prev_x = 0
        self.prev_y = 0
        self.ls_time = None
        self.shk_cell = 66
        self.CybZ_03_act = False

        self.ut_count = 0
        self.ut_place = 47,48,38,39

        self.system_services = (
            self.ap_hash_controller,
            self.pos_rob_controller,
            self.driver_controller,
            self.zonya,
            self.speed_check,
            self.magnit,
            self.shkaf_oshibka,
            self.oshibka_false
        )
        
        
       
    

    def trusted_code(self):
        time.sleep(3)
        self.ap_hash = self.get_ap_code_hash()
        self.evacuation = False

        self.system_start()

    def system_start(self):
        for service in self.system_services:
            service_thread = threading.Thread(target = service)
            service_thread.start()
        
    def ap_hash_controller(self):
        """
        Функция для контроля аутентичности кода автопитоа CybP_01.
        """
        while True:
            cur = self.get_ap_code_hash()
            code_state = cur != self.ap_hash
            if code_state and not self.CybP_01_act:
                self.lg.log("CybP_01: Высокий уровень,запрос на перезагрузку прошивки автопилота")
                self.set_ap_force_reset()
                self.CybP_01_act = True
            if not code_state and self.CybP_01_act:
                self.lg.log('CybP_01: Высокий уровень,перезагрузка прошивки автопилота успешно')
                self.CybP_01_act = False
            time.sleep(1)
        
    def driver_controller(self):
        while True:
            drive_data = self.get_drive_data()
            raw = drive_data
            d_data =  raw.get('drive_data')
            if not d_data: continue
    
            for i in range(0, 4):
                unit_data = d_data[i]
                if not unit_data:continue
                n_mock_data = unit_data['data']
                unit_serial = unit_data['serial']
                received = unit_data['last_received_from']
                n_str = self.calc_hash((str(unit_serial + n_mock_data)))
                if n_str != received:
                    self.lg.log("CybP_03")
                    self.set_drive_force_reset(unit_data)
            time.sleep(0.3)
            
    def pos_rob_controller(self):
        while True:
            info = self.get_robot_status()
            self.pos_x = info.get('position_x')
            self.pos_y = info.get('position_y')
            time.sleep(0.2)

    def calc_hash(self,msg: str):
        c_hash = crc8.crc8()
        c_hash.update(bytes(msg.encode('utf-8')))
        hash_h = c_hash.hexdigest()
        return hash_h

    def get_x(self,point):
        x = (point - 1) % 9 + 1
        return x * 0.4 - 0.2

    def get_y(self,point):
        y = (point - 1) // 9 + 1
        return y * 0.4 - 0.2

    def check_rob_in_cell(self,num_cell) -> bool:
        self.cell_x = self.get_x(num_cell)
        self.cell_y = self.get_y(num_cell)
        if abs(self.pos_x - self.cell_x) < self.cell_centre_raduis and abs(self.pos_y - self.cell_y) < self.cell_centre_raduis:
            return True
        else:
            return False
    def zonya(self):
        while True:
            zona = self.get_service_zones()
            zones = zona.get('service_zones')
            detected = False
    
            if zones:
                for i in zones:
                    rob_status = self.check_rob_in_cell(i)
                   
                    if rob_status:
                        detected = True
                        break
            if detected:
                if not self.evacuation:
                    self.evacuation = True
                    self.lg.log("CybP_04 stop")
                    self.set_emergency_stop(True)
    
            else:
                if self.evacuation:
                    self.evacuation = False
                    self.lg.log("CybP_04 cont")
                    self.set_emergency_stop(False)
            time.sleep(1)

    def speed_check(self):
        while True:
            cur = time.time()
            t = cur - self.next_t

            t = round(t,2)
            
            x = abs(self.pos_x - self.prev_x)
            y = abs(self.pos_y - self.prev_y)
            S = (round(x,2) + round(y,2))
            V = S/t
            
            if 0.02 < V < 0.1:
                if self.ls_time is None:
                    self.ls_time = cur
                elif (cur - self.ls_time) > 4.0:
                    self.lg.log("CybZ_01")
                    self.set_speed_controller_reset()
                    self.ls_time = None
            else:
                self.ls_time = None
    
            self.prev_x = self.pos_x
            self.prev_y = self.pos_y
            self.next_t = time.time()
            time.sleep(1)
            
    def magnit(self):
        while not self.CybZ_02:
            raw = self.get_system_messages()
            sms = raw.get('system_messages')
            if sms and 'неполадки' in str(sms[-1]):
                self.CybZ_02 = True
                self.lg.log("CybZ_02")
                self.set_grip_lock(True)
            time.sleep(2)


    def shkaf_oshibka(self):
        while True:
            if self.check_rob_in_cell(self.shk_cell) and not self.CybZ_03_act:
                temp = self.get_control_color_temperature()
                print(temp)
                try:
                    col = temp.get('control_color_temp')
                    self.send_message_to_ap(col)
                    self.CybZ_03_act = True
                except Exception as e:
                    self.lg.log(e)
            time.sleep(0.5)

    def oshibka_false(self):
        while True:
            val = False
            status = False
            if self.ut_count < 4:
                if self.check_rob_in_cell(self.ut_place[self.ut_count]):
                    count = 0
                    for i in range(10):
                        raw = self.get_reserved_cell_pollution_status()
                        
                        try:
                            val = raw.get('state')
                        except:
                            val = False
                        if val:
                            count += 1
                        if count >= 7:
                            status = True
                    if status:
                        self.send_message_to_ap(self.ut_count + 1)
                    self.ut_count += 1
            time.sleep(0.6)
            
    def make_next_short_message(self, prev_message: str):
        
        hash_h = None

        if self.short_massage is None:
            self.short_massage = prev_message
            return self.short_massage
        hash_h = self.calc_hash(self.short_massage)
        
        if hash_h != prev_message:
            if self.CybP_02_status == 0:
                self.CybP_02_status += 1
                start_time = datetime.datetime.now()
                self.lg.log(f"CybP_02:Уровень – средний, потеря связи с диспетчером:{start_time}")
                return self.short_massage

        if self.CybP_02_status == 1:
            end_time = datetime.datetime.now()
            self.lg.log(f"CybP_02:Уровень – средний,связь с диспетчером востановлена:{end_time}")
            self.CybP_02_status += 1

        return self.short_massage

        # Ваш код для создания коротких сообщений.
        # Это - функция-заготовка.
        # Удалите определение "pass" для активации функции, а также создайте код, генерирующий реальный some_message, который данная функция должна возвращать.

    # Вы можете добавить ваши собственные функции, которые будут вызываться trusted_code (подсказка - если вы вынесете функционал в отдельную функцию, его проще будет оценить в случае разногласий).
    # Автоматически вызываются только функции trusted_code (один раз при вызове start_module()) и make_next_short_message (каждую секунду).
```

Для запуска заезда вам необходимо вызвать функцию `start_module()`, передав им ваши UserMissionHandler и UserTrustedHandler.

В случае физического робота, комангда на старт заезда подаётся при помощи нажатии на большую зелёную кнопку, располагающуюся на пульте управления роботом (вариант через команду так же доступен, но с кнопкой всё же эпичнее).

```python
start_module(UserMissionHandler, UserTrustedHandler)
```

Завершить заезд можно при помощи команды `/stop`, подаваемой в АСО.

Для физического робота доступен аналог зелёной кнопки, начинающей заезд - Красная кнопка, завершающая заезд.

При желании начать новый заезд вам необходимо ввести команду `/reset`, которая сбросит инциализаторы АСО. Затем необходимо перезапустить ЦД (закрыть и открыть заново).
